```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
```

# Linear Regression Using Sklearns Inbuilt Dataset

```python
def mse(y_test, predictions):
        return np.mean((y_test - predictions)**2)

class LinearRegression:
    
    def __init__(self, lr = 0.01, n_iters = 1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
        
    #Training the model
    def fit(self, X ,y):
        n_samples, n_features = X.shape 
        #returns (rows (n_samples), columns (n_features))
        
        self.bias = 0
        self.weights = np.zeros(n_features) 
        #initalize multiple weights for multiple parameters(data)
        
        
        for i in range(self.n_iters):
            #Instead of predicting each sample one-by-one, we form an array of
            #samples and do predictions for all samples at the same time        
            #Formula : y_pred = wX + b
            y_pred = np.dot(X, self.weights) + self.bias

            #Calculate Gradients (2* is adjusted in the learning rate parameter)
            #np.dot includes both matrix multiplication and summation
            dw = (1/n_samples) * np.dot(X.T,(y_pred - y))
            db = (1/n_samples) * np.sum(y_pred - y)

            #Update the weights and bias
            self.weights = self.weights - self.lr * dw
            self.bias = self.bias - self.lr * db
            
            #Model is trained of completing n_iters iterations
            
    def predict(self, X):
        y_pred = np.dot(X, self.weights) + self.bias
        return y_pred
```

```python
from sklearn import datasets

X,y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], y, color="blue", s = 10, label='Data Points')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.show()

model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)

def mse(y_test, predictions):
        return np.mean((y_test - predictions)**2)

mse = mse(y_test, predictions)
print("Mean Squared Error:",mse)

y_pred_line = model.predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X_train, y_train, color = 'blue', s=10, label='Training Data')
plt.scatter(X_test, y_test, color ='red', s=10, label='Testing Data')
plt.plot(X, y_pred_line, color='black', linewidth=2, label='Predictions')
plt.legend()
plt.show()
```
